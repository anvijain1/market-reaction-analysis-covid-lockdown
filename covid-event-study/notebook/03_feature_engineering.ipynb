```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
BASE_DIR = Path.cwd().parent
prices = pd.read_csv(
    BASE_DIR / "data" / "processed" / "close_prices.csv",
    index_col=0,
    parse_dates=True
)

prices.head()
```

```python
returns = np.log(prices / prices.shift(1)) #log return
returns = returns.dropna()

returns.head()
```

```python
returns.shape
```

```python
returns.isna().sum()
```

```python
returns.describe()
```

```python
rolling_volatility = returns.rolling(window=21).std()
rolling_volatility.head()
```

```python
PROCESSED_DIR = BASE_DIR / "data" / "processed"

returns.to_csv(PROCESSED_DIR / "daily_returns.csv")
rolling_volatility.to_csv(PROCESSED_DIR / "rolling_volatility.csv")
```
