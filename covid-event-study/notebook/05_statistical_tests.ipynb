```python
import pandas as pd
import numpy as np
from scipy import stats
from pathlib import Path
```

```python
BASE_DIR = Path.cwd().parent

abnormal_returns = pd.read_csv(
    BASE_DIR / "data" / "processed" / "abnormal_returns.csv",
    index_col=0,
    parse_dates=True
)

abnormal_returns.head()
```

```python
results = {}

for stock in abnormal_returns.columns:
    data = abnormal_returns[stock]
    
    t_stat, p_value = stats.ttest_1samp(data, 0)
    
    results[stock] = {
        "Mean_AR": data.mean(),
        "T_stat": t_stat,
        "P_value": p_value
    }

results_df = pd.DataFrame(results).T
results_df
```
