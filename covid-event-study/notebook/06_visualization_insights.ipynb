```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

plt.style.use("seaborn-v0_8")
```

```python
BASE_DIR = Path.cwd().parent

abnormal_returns = pd.read_csv(
    BASE_DIR / "data" / "processed" / "abnormal_returns.csv",
    index_col=0,
    parse_dates=True
)

CAR = pd.read_csv(
    BASE_DIR / "data" / "processed" / "CAR.csv",
    index_col=0,
    parse_dates=True
)

volatility = pd.read_csv(
    BASE_DIR / "data" / "processed" / "rolling_volatility.csv",
    index_col=0,
    parse_dates=True
)
```

```python
plt.figure(figsize=(12,6))

for stock in CAR.columns:
    plt.plot(CAR.index, CAR[stock], label=stock)

plt.axvline(pd.to_datetime("2020-03-24"), color="red", linestyle="--", label="Lockdown")

plt.title("Cumulative Abnormal Returns (COVID Event Window)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))

for stock in abnormal_returns.columns:
    plt.plot(abnormal_returns.index, abnormal_returns[stock], label=stock)

plt.axhline(0, color="black", linestyle="--")
plt.axvline(pd.to_datetime("2020-03-24"), color="red", linestyle="--")

plt.title("Daily Abnormal Returns Around Lockdown")
plt.legend()
plt.show()
```
